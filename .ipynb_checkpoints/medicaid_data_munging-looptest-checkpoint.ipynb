```python
import pandas as pd
from datetime import datetime
import os
import pymongo
```

```python
appended_data = []
path = os.path.join("Data","Raw","Medicaid_by_Year","2019")
for WorkingFile in os.listdir(path):
        #identifies the list of files to be grouped together from designated folder
        print(WorkingFile)
        # Import the csv file by joining the path directory and file name together
        file = pd.read_csv(os.path.join(path, WorkingFile))
        # Load the csv file as a dataframe
        df = pd.DataFrame(file)
        # Code from Medicaid Data Munging to Clean Up Files
        print(df.count())
        print(df.isnull().any())
        print(df.loc[df.isnull().any(axis=1)])
        df_2019 = df.fillna(value=0)
        print(df_2019.count())
        print(df.columns.values)
        df_2019 = df_2019.drop(["Preliminary_Updated", 
                              "Final_Report",
                              "New Applications Submitted to Medicaid and CHIP Agencies",
                              "New Applications Submitted to Medicaid and CHIP Agencies – footnotes",
                              "Applications for Financial Assistance Submitted to the State Based Marketplace",
                              "Applications for Financial Assistance Submitted to the State Based Marketplace – footnotes",
                              "Total Applications for Financial Assistance Submitted at State Level – footnotes",
                              "Individuals Determined Eligible for Medicaid at Application - footnotes",
                              "Individuals Determined Eligible for CHIP at Application - footnotes",
                              "Total Medicaid and CHIP Determinations - Footnotes",
                              "Medicaid and CHIP Child Enrollment",
                              "Medicaid and CHIP Child Enrollment - footnotes",
                              "Total Medicaid and CHIP Enrollment - footnotes",
                              "New Georeferenced Column","Total Medicaid Enrollment - footnotes",
                              "Total CHIP Enrollment - footnotes"]
                             ,axis=1)
        df_2019.rename(columns = {'State Abbreviation' : 'State_Abbr', 
                             'State Name' : 'State', 
                             'State Expanded Medicaid' : 'Medicaid_Expanded', 
                             'Total Applications for Financial Assistance Submitted at State Level' : 'Total_Financial_Assistance_Apps_Submitted', 
                             'Individuals Determined Eligible for Medicaid at Application':'Medicaid_Indiv_Elig_Applied',
                             'Individuals Determined Eligible for CHIP at Application':'CHIP_Indiv_Elig_Applied',
                             'Total Medicaid and Chip Determinations':'Total_Determinations',
                             'Total Medicaid and CHIP Enrollment':'Total_Enrollment',
                             'Total Medicaid Enrollment':'Tot_Medicaid_Enroll',
                             'Total CHIP Enrollment':'Tot_CHIP_Enroll'}, 
                        inplace = True)
        df_2019['Report Date'] = pd.to_datetime(df_2019['Report Date'])
        
        def mapper(month):
           return month.strftime('%b') 

        df_2019['Month'] = df_2019['Report Date'].apply(mapper)
        df_2019['Year'] = df_2019['Report Date'].dt.year
        df_2019 = df_2019.drop('Report Date',axis=1)
        # append to list
        appended_data.append(df_2019)
#convert to dataframe
appended_data = pd.concat(appended_data)
```
