### Import dependecies, check .csv file directory, import and create dataframe 

```python
import pandas as pd
from datetime import datetime
import os
import pymongo
```

### Import and Append Annual Summary Data

```python
file_19 = pd.read_csv(os.path.join("Data", "Clean", "2019_Full.csv"))
file_18 = pd.read_csv(os.path.join("Data", "Clean", "2019_Full.csv"))
file_17 = pd.read_csv(os.path.join("Data", "Clean", "2019_Full.csv"))
#file_16 = pd.read_csv(os.path.join("Data", "Clean", "2019_Summary.csv"))
#file_15 = pd.read_csv(os.path.join("Data", "Clean", "2019_Summary.csv"))
#file_14 = pd.read_csv(os.path.join("Data", "Clean", "2019_Summary.csv"))
```

```python
df_19 = pd.DataFrame(file_19)
df_19.head()
```

```python
df_18 = pd.DataFrame(file_18)
df_18['Year']='2018'
df_18.head()
```

```python
df_17 = pd.DataFrame(file_17)
df_17['Year']='2017'
df_17.head()
```

```python
all_years = df_19.append(df_18)
```

```python
all_years = all_years.append(df_17)
```

```python
all_years.head(100)
```

```python
all_years.to_csv(os.path.join("Data","Clean","All_Years.csv"))
```

```python
all_years_dict = all_years.to_dict(orient="records")
```

```python
all_years_dict
```

```python
nested_list = []
nested_dict = {}
```

```python
state_list = []
for row in all_years_dict:
    state_list.append(row["State"])
```

```python
year_list = []
for row in all_years_dict:
    year_list.append(row["Year"])
```

```python
for i in range(len(state_list)):
    nested_dict = {"State": state_list[i], "Year": year_list[i], "Data": all_years_dict[i]}
    
    nested_list.append(nested_dict)
```

```python
nested_list
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.ACAData_DB
collection = db.ACAStateData
```

```python
for data in nested_list:
    collection.insert_one(data)
```
