```python
import pandas as pd
from datetime import datetime
import os
import numpy as np
```

```python
file = pd.read_csv(os.path.join("..","..","Data","Raw","America_Health_Rankings","2017-Annual.csv"))
print(file)
```

```python
df2017 = pd.DataFrame(file)
df2017.head()
```

```python
df2017.count()
```

```python
df2017['Measure Name'].unique()
```

```python
#'Below poverty threshold','Cancer Deaths','Cardiovascular Deaths','Cholesterol Check','Diabetes','Drug Deaths',
```

```python
Cardiovascular_Deaths=df2017[df2017['Measure Name']=='Cardiovascular Deaths']
```

```python
Cardiovascular_Deaths.head()
```

```python
Cardiovascular_Deaths.isnull().any()
```

```python
Cardiovascular_Deaths.loc[Cardiovascular_Deaths.isnull().any(axis=1)]
```

```python
Cardiovascular_Deaths = Cardiovascular_Deaths.fillna(value=0)
Cardiovascular_Deaths
```

```python
Cardiovascular_Deaths.count()
```

```python
list(Cardiovascular_Deaths.columns.values)
```

```python
Cardiovascular_Deaths = Cardiovascular_Deaths[["State Name",
            "Edition", 
            "Measure Name",
            "Rank",
            "Value"]]
```

```python
Cardiovascular_Deaths.head()
```

```python
Cardiovascular_Deaths.rename(columns = {'Rank' : 'Cardiovascular_Deaths_Rank', 
                             'Value' : 'Cardiovascular_Deaths_Value',
                             'State Name': 'State',
                             'Edition': 'Year'}, 
                  inplace = True)
```

```python
Cardiovascular_Deaths.head()
```

```python
Cardiovascular_Deaths = Cardiovascular_Deaths.drop('Measure Name',axis=1)
```

```python
Cardiovascular_Deaths.head()
```
