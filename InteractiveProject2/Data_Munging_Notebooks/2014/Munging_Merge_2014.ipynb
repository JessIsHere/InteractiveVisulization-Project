# Merging Annual Medicaid Summary Data to Annual AHR Data Metrics

### Import dependecies, check .csv file directory, import and create dataframe 

```python
import pandas as pd
from datetime import datetime
import os
import pymongo
```

### Import Medicaid Summary Data

```python
medicaid_path = os.path.join("Data", "Clean", "2014_Summary.csv")
medicaid_file = pd.read_csv(medicaid_path)
```

```python
medicaid14_df = pd.DataFrame(medicaid_file)
medicaid14_df.head()
```

```python
i = 0
path = os.path.join("Data","Raw","America_Health_Rankings","2014 Midstep")
for WorkingFile in os.listdir(path):
        #identifies the list of files to be grouped together from designated folder
        print(WorkingFile)
        # Import the csv file by joining the path directory and file name together
        file = pd.read_csv(os.path.join(path, WorkingFile))
        # Load the csv file as a dataframe
        df = pd.DataFrame(file)
        # merge files use if statement to initalize dataframe that will then be final df called once loops complete
        if i==0:
            df_2014 = pd.merge(medicaid14_df, df, on=['State','Year'], how='outer')
        else:
            df_2014 = pd.merge(df_2014, df, on=['State'], how='outer')
            
        i+=1
```

```python
df_2014
```

```python
df_2014 = df_2014.drop(['Year_x','Year_y','Month'],axis=1)
```

```python
df_2014
```

```python
df_2014['Year']='2014'
```

```python
 df_2014['Year'] = pd.to_datetime(df_2014['Year'])
```

```python
df_2014['Year'] = df_2014['Year'].dt.year
```

```python
df_2014 = df_2014[df_2014['State'] != 'District of Columbia']
```

```python
df_2014 = df_2014[df_2014['State']!='United States']
```
